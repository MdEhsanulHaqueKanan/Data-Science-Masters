Q1. Explain GET and POST methods.

Answer: GET is an HTTP method used to retrieve data from a server. It is used for reading and fetching information, and it should not modify any data on the server.

POST is an HTTP method used to send data to a server to create or modify a resource. It is used for submitting data, such as form submissions, and can modify data on the server.

Q2. Why is request used in Flask?

Answer: The request object in Flask is used to access incoming request data. It allows the Flask application to retrieve information from the client's request, such as form data, query parameters, headers, and JSON payloads. It provides a convenient way to handle and process the data sent by the client.

Q3. Why is redirect() used in Flask?

Answer: The redirect() function in Flask is used to redirect the user's browser to a different route or URL. It is commonly used after processing a form submission or performing certain actions, allowing the server to redirect the user to a different page or route for further processing or displaying a success message.

Q4. What are templates in Flask? Why is the render_template() function used?

Answer: Templates in Flask are files that contain HTML code with placeholders for dynamic content. They allow you to separate the presentation logic from the application logic. 

The render_template() function is used to render these templates and generate HTML pages dynamically. It takes the name of the template file as an argument and replaces the placeholders with actual data, allowing the Flask application to generate dynamic web pages to be served to the client.

Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook

Answer: Here is the code of a simple API:

```python
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)


@app.route('/')
def home_page():
    return render_template('index.html')


@app.route('/math',methods=['POST'])
def math_ops():
    if(request.method == 'POST'):
        ops = request.form['operation']
        num1 = int(request.form['num1'])
        num2 = int(request.form['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return render_template('results.html' , result = result)



@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

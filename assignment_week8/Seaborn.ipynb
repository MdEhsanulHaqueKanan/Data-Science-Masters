Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer: Here are five plots we can plot using the Seaborn library

1. Scatter Plot: 
Seaborn's scatter plot is used to visualize the relationship between two continuous variables. It helps in identifying patterns, clusters, or trends in the data.

2. Line Plot: 
Seaborn can be used to create line plots, which are useful for displaying trends or patterns over time or any continuous variable.

3. Bar Plot: 
Seaborn's bar plot is helpful for comparing categorical data, showing the distribution or count of each category, and identifying the differences between groups.

4. Histogram: 
Seaborn provides histogram plots to visualize the distribution of a single variable, showing the frequency or count of values within each bin.

5. Heatmap: 
Seaborn's heatmap is used to display a matrix-like dataset, with each cell colored based on its value. It is effective in visualizing correlations, patterns, or relationships between variables in a dataset.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

Answer: Here is the code:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Plot line plot for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and labels
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

Answer: Here is the code:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create subplots for box plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Box plot for 'age' by 'pclass'
sns.boxplot(ax=axes[0], data=titanic, x='pclass', y='age')
axes[0].set_title('Age by Pclass')

# Box plot for 'fare' by 'pclass'
sns.boxplot(ax=axes[1], data=titanic, x='pclass', y='fare')
axes[1].set_title('Fare by Pclass')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Answer: Here is the code:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut' column
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

# Set plot title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Answer: Here is the code: 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Reshape the data into a pivot table format
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Set plot title and labels
plt.title("Passenger Count Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```

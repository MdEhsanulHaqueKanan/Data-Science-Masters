Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer: Matplotlib is a popular data visualization library in Python. 

It is used to create high-quality plots, charts, and figures for data analysis and presentation. Matplotlib provides a wide range of plot types and customization options, allowing users to create visually appealing and informative visualizations. It is extensively used in fields such as data science, scientific research, engineering, and finance to analyze and communicate data effectively.

The Pyplot module of Matplotlib provides various types of plots for data visualization. Five commonly used plots that can be created using Pyplot are:

1. Line plot: 
Displays data points connected by straight lines, useful for showing trends or patterns over time or continuous variables.
2. Scatter plot: 
Represents individual data points as markers on a 2D plane, helpful for visualizing the relationship between two variables.
3. Bar plot: 
Shows categorical data with rectangular bars, suitable for comparing different categories or groups.
4. Histogram: 
Illustrates the distribution of a dataset by dividing it into bins and displaying the frequency or count of values within each bin.
5. Pie chart: 
Represents data as slices of a circular pie, providing a visual representation of the proportion or percentage of each category within a dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Note: Also add title, xlabel, and ylabel to the plot.

Answer: A scatter plot is a type of plot that displays individual data points as markers on a 2D plane. It is useful for visualizing the relationship between two variables and identifying patterns or correlations.

Here is the code to plot of scatter plot:

```python
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y)

# Adding title, xlabel, and ylabel
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Display the plot
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer: The subplot() function in Matplotlib is used to create multiple plots within a single figure.
Here is the code to draw four line plots:

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line 4")

# Adjusting subplot spacing
plt.tight_layout()

# Display the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Answer: A bar plot is a type of plot that represents categorical data with rectangular bars. 

It is used to compare and visualize the values of different categories or groups. The length or height of each bar corresponds to the value of the category it represents.

Here is the code to plot a bar plot and a horizontal bar plot:

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 4))  # Set the figure size
plt.bar(company, profit)  # Plotting the bar plot
plt.title("Profit by Company")  # Adding title
plt.xlabel("Company")  # Adding x-label
plt.ylabel("Profit")  # Adding y-label
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 4))  # Set the figure size
plt.barh(company, profit)  # Plotting the horizontal bar plot
plt.title("Profit by Company")  # Adding title
plt.xlabel("Profit")  # Adding x-label
plt.ylabel("Company")  # Adding y-label
plt.show()
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Answer: A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values of a dataset, along with any outliers.

A box plot is used to identify the central tendency, spread, and skewness of a dataset. It allows for easy comparison of multiple datasets or groups and helps in detecting potential outliers or extreme values.

Here is the code to plot a box plot:

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 4))  # Set the figure size
plt.boxplot(data)  # Plotting the box plot
plt.title("Box Plot")  # Adding title
plt.xticks([1, 2], ["Box 1", "Box 2"])  # Adding x-axis labels
plt.ylabel("Values")  # Adding y-label
plt.show()
```
